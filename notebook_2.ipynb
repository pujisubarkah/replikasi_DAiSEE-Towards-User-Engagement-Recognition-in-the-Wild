```python
subject = "110001"

subject_path = os.path.join(TRAIN_PATH, subject)

clip_dirs = sorted([
    d for d in os.listdir(subject_path)
    if os.path.isdir(os.path.join(subject_path, d))
])

selected_clip = clip_dirs[0]

clip_path = os.path.join(subject_path, selected_clip)

videos = sorted([
    f for f in os.listdir(clip_path)
    if f.endswith(".avi")
])

video_path = os.path.join(clip_path, videos[0])

print(video_path)
```

```python
#load video
def load_video(video_path):

    cap = cv2.VideoCapture(video_path)

    frames = []

    while True:

        ret, frame = cap.read()

        if not ret:
            break

        frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

        frames.append(frame)

    cap.release()

    return frames
```

```python
frames = load_video(video_path)

print("Jumlah frame:", len(frames))
```

```python
def uniform_sampling(frames, num_frames=30):

    indices = np.linspace(
        0,
        len(frames)-1,
        num_frames,
        dtype=int
    )

    sampled = [frames[i] for i in indices]

    return sampled
```

```python
sampled_frames = uniform_sampling(frames)

print(len(sampled_frames))
```

```python
def resize_frames(frames, size=(224,224)):

    return [
        cv2.resize(frame, size)
        for frame in frames
    ]
```

```python
def normalize_frames(frames):

    frames = np.array(frames, dtype=np.float32)

    frames /= 255.0

    return frames
```

```python
def preprocess_video(video_path):

    frames = load_video(video_path)

    frames = uniform_sampling(frames)

    frames = resize_frames(frames)

    frames = normalize_frames(frames)

    return frames
```

```python
video = preprocess_video(video_path)

print(video.shape)
```
